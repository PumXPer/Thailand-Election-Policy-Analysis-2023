```python
import bs4
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pythainlp.tokenize as tokenize
```

# Move Forward

```python
title_list = []
problem_list = []
solution_list = []
```

```python
for n in range (1,314):
  data = requests.get('https://election66.moveforwardparty.org/policy/detail/policy_'+str(n))
  soup = bs4.BeautifulSoup(data.text)

  title = soup.find('div',{'class':'sc-ceb6e8f9-0 MSOum SectionTitle'})
  if title is not None:
    title_list.append(title.text)

  problem = soup.find('div',{'class':'problem'})
  if problem is not None:
    problem_list.append(problem.text)

  solution = soup.find('div',{'class':'solution'})
  if solution is not None:
    solution_list.append(solution.text)

  print(n)
```

```python
table_move_forward = pd.DataFrame([title_list,problem_list,solution_list]).T
table_move_forward.columns = ['title_list','problem_list','solution_list']
table_move_forward.set_index('title_list')
```

```python
table_move_forward.to_csv('move_forward.csv')
```

```python
mdf = pd.read_csv('move_forward.csv')
```

```python
solution_text = mdf['solution_list'].values
```

```python
text = ' '.join([solution_text[0] for word in solution_text])
```

```python
tokens = tokenize.word_tokenize(text, engine='newmm')
```

```python
# tokens
```

```python
stopwords = ['ไม่','ใน','การ','ของ','มา','จาก','ใช่','ผู้','ให้','ใหม่','โดย','มี','เพื่อ','ปิด','และ','ได้','ที่','เพิ่ม','อย่าง','น้อย','ง่าย','เพิ่ม','ถูก','ห้าม','กับ'
,'ทุก']
```

```python
mtext = ' '.join(tokens)
```

```python
mtext
```

```python
wordcloud = WordCloud(
regexp='[ก-๙]+',
font_path='THSarabunNew.ttf',

stopwords=stopwords,
width=2000, height=1000,
prefer_horizontal=0.9,
colormap='viridis',
background_color = 'white').generate(mtext)

plt.figure(figsize = (10, 9))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
```

# ptp

```python
data = requests.get('https://ptp.or.th/')
soup = bs4.BeautifulSoup(data.text)
```

```python
href_list = []
for div in soup.find_all('h3', class_='elementor-heading-title elementor-size-default'):
    a = div.find('a')
    if a is not None:
        href_list.append(a['href'])
```

```python
title_list_ptp = []
solution_list_ptp = []
```

```python
for ptp_url in href_list:
  url = requests.get(ptp_url)
  soup = bs4.BeautifulSoup(url.text)

  title_ptp = soup.find('h3',{'class':'elementor-heading-title elementor-size-xxl'})
  if title_ptp is not None:
    title_list_ptp.append(title_ptp.text)
    print(title_ptp.text)

  solution_ptps = soup.find_all('div',{'class':'elementor-text-editor elementor-clearfix'})
  if solution_ptps is not None:
    # solution_list_ptp.append(solution_ptp.text)
    solution_ptp = [item.text.strip() for item in solution_ptps]
    solution_list_ptp.append(solution_ptp)

  print(ptp_url)
```

```python
table_ptp = pd.DataFrame([title_list_ptp,solution_list_ptp]).T
table_ptp.columns = ['title_list_ptp','solution_list_ptp']
table_ptp.set_index('title_list_ptp')
```

```python
table_ptp.to_csv('ptp.csv')
```
